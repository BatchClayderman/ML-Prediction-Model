```python
data=pd.read_csv('result/final_tableV10.csv')
data['first_careunit']=data['first_careunit'].map({'MICU':0, 'CCU':1, 'NICU':2, 'SICU':3, 'TSICU':4, 'CSRU':5})
data['ethnicity']=data['ethnicity'].map({'WHITE':1, 'ASIAN':2, 'BLACK/AFRICAN AMERICAN':3,
       'UNKNOWN/NOT SPECIFIED':0, 'HISPANIC OR LATINO':1,
       'PATIENT DECLINED TO ANSWER':0, 'MULTI RACE ETHNICITY':4, 'OTHER':0,
       'ASIAN - ASIAN INDIAN':2, 'HISPANIC/LATINO - HONDURAN':1,
       'BLACK/HAITIAN':3, 'UNABLE TO OBTAIN':0, 'ASIAN - VIETNAMESE':2,
       'WHITE - BRAZILIAN':1, 'WHITE - RUSSIAN':1, 'ASIAN - KOREAN':2,
       'ASIAN - CHINESE':2, 'WHITE - OTHER EUROPEAN':1,
       'HISPANIC/LATINO - PUERTO RICAN':1, 'ASIAN - CAMBODIAN':2})
```

```python
data['group']=0
data.loc[data[(data['gender']==1) & (data['uric_acid.csv']>=7)].index, 'group']=1
data.loc[data[(data['gender']==0) & (data['uric_acid.csv']>=6)].index, 'group']=1
```

```python
data=pd.read_csv('result/finaltable.csv')
data.drop(['Unnamed: 0'], axis=1,inplace=True)
data['low_group']=0
data['high_group']=0
data.loc[data[data['group']==1].index,'high_group']=1
data.loc[data[data['group']==0].index,'low_group']=1
```

```python
data['age_class']=0
data['BMI_class']=0
```

```python
data.loc[data[data['age']>=65].index, 'age_class']=1
```

```python
data.loc[data[(data['BMI']>=18.5)&(data['BMI']<=24.9)].index, 'BMI_class']=1
data.loc[data[(data['BMI']>=24.9)].index, 'BMI_class']=2
```
