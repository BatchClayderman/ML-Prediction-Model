```python
import sys
sys.path.append("..")
import myFunction as mf
import os
import pandas as  pd
import myFunction as mf
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import auc, roc_curve
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import warnings
from sklearn.externals import joblib
#import joblib
from xgboost.sklearn import XGBClassifier
import random
from sklearn.utils import shuffle
from scipy import stats
import myFunction as mf
from sklearn.linear_model import Lasso
from sklearn.neural_network import MLPClassifier
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None) #show all columns
pd.set_option('max_colwidth',None)
```

```python
data=pd.read_csv(os.path.dirname(os.getcwd())+'/result/original data.csv')
```

```python
data.columns
```

```python
train_variables=['ethnicity', 'gender', 'los','age', 'acute pancreatitis.csv', 
                 'Chronic liver disease.csv', 'CKD.csv',
                'COPD.csv', 'coronary heart disease.csv', 'diabetes.csv',
       'has_surgery.csv', 'heart failure.csv', 'hypertension.csv',
       'hypotension.csv', 'kidney transplantation.csv', 'Malignancy.csv',
       'rhabdomyolysis.csv', 'Albumin.csv_mean', 'Chloride.csv_mean',
       'creatinine.csv_mean', 'Hemoglobin.csv_mean', 'platelet.csv_mean',
       'potassium.csv_mean', 'RDW.csv_mean', 'sodium.csv_mean',
       'TBIL.csv_mean', 'urea nitrogen.csv_mean', 'WBC.csv_mean','aki_1',
       'aki_2', 'aki_3', 'BMI', 'ratio']
outcome_variables=['365_mortality']
```

```python
split_percentage=0.8
sub_data=data[list(train_variables)+outcome_variables]
train=sub_data[:int(len(data)*split_percentage)]
test=sub_data[int(len(data)*split_percentage):]
```

```python
original_train_variables=train_variables

report=[]
time=0
min_child_weight_list=[0.2,8,20]

for mark0, apha in enumerate([0.01]):
    lasso = Lasso(alpha=apha)
    lasso.fit(train[original_train_variables], train[outcome_variables])
    lasso.score(test[original_train_variables], test[outcome_variables])
    result=[]
    result.append(original_train_variables)
    result.append(list(lasso.coef_))
    train_variables=[]
    for i in range(len(result[0])):
        if result[1][i]!=0:
            train_variables.append(result[0][i])

    for mark1,scale_pos_weight in enumerate([30,3000,7000]):
        for mark2,n_estimators in enumerate([3,30,60]):
            for mark3,max_depth in enumerate([3,30,90]):
                for mark4,reg_lambda in enumerate([0.1,1,10]):
                    for mark5,gamma in enumerate([0.1,1,10]):
                        for mark6, min_child_weight in enumerate(min_child_weight_list):
                            for mark7, subsample in enumerate([0.3,0.6,1]):
                                for mark8, max_delta_step in enumerate([0,3]):
                                    for count,i in enumerate(outcome_variables):
                                        time+=1
                                        parameters=[{'n_estimators': [n_estimators], 
                                                    'max_depth': [max_depth],
                                                    'reg_lambda':[reg_lambda],
                                                    'gamma':[gamma],
                                                    'scale_pos_weight':[scale_pos_weight]}]
                                        model=XGBClassifier(n_jobs=-1, n_estimators=n_estimators, max_depth=max_depth,
                                                                      reg_lambda=reg_lambda,gamma=gamma,scale_pos_weight=scale_pos_weight,
                                                                       min_child_weight=min_child_weight,
                                                                       subsample=subsample,max_delta_step=max_delta_step)
                                        model.fit(train[train_variables], train[i])
                                        y_score=model.predict_proba(test[train_variables])[:,1]
                                        fpr, tpr, _ = roc_curve(test[i], y_score)
                                        AUROC=auc(fpr, tpr)
                                        trainS=model.score(train[train_variables], train[i])
                                        testS=model.score(test[train_variables], test[i])
                                        f1S=metrics.f1_score(test[i], model.predict(test[train_variables]))
                                        preS=metrics.precision_score(test[i], model.predict(test[train_variables]))
                                        recS=metrics.recall_score(test[i], model.predict(test[train_variables]))
                                        AUROC=auc(fpr, tpr)
                                        report.append([trainS, testS, f1S, preS,recS,AUROC, i,scale_pos_weight,
                                                      n_estimators,max_depth,reg_lambda,gamma,min_child_weight,
                                                      subsample,max_delta_step,apha,result,len(train_variables)/33])
                                        if time%100==0:
                                            print(i,':', time)
                                        #print(trainS,'-',testS,'-',f1S,'-',preS,'-',recS,'-',AUROC,'-',model.best_params_)
                                        #print('---------------------------------')
                                    res=pd.DataFrame(report)
                                    res.to_csv('C:/Programs/Python Project/result/report.csv')
```
