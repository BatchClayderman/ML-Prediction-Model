```python
import psycopg2
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_columns', None) #show all columns
pd.set_option('max_colwidth',None)
```

```python
surgery=pd.read_csv(os.path.dirname(os.getcwd())+'/data/original data/surgery.csv')
```

```python
def search_table(s='*', f='admissions', w='', like='', download_as=''):
    temp=s.lower()
    for i in temp:
        if i>"z" or i<"a":
            assert "TypeError"
    del temp
    conn = psycopg2.connect(database="mimic", user="postgres",
                            password="pandora", host="127.0.0.1", port="5432")
    cursor = conn.cursor()
    if (w!='')&(like!=''):
        cursor.execute("select " + s + " from mimiciii." + f +' where '+ w +
                       ' like ' + "'%"+ like +"%'")
    elif (w!='')&(like=='') :
        cursor.execute("select " + s + " from mimiciii." + f +' where '+ w)
    else:
        cursor.execute("select " + s + " from mimiciii." + f)
    table = cursor.fetchall()
    if(len(table)!=0):
        colnames = [desc[0] for desc in cursor.description]
        columns = cursor.fetchall()
        cursor.close()
        conn.close()
        columns=pd.DataFrame(columns)
        table=pd.DataFrame(table)
        table.columns=colnames
    if(download_as!=''):
        table.to_csv(os.path.dirname(os.getcwd())+'/data/download data/'+download_as+'.csv', index=False)
    return table
```

```python
def searchSurgery(surgery_name, download_as=''):
    result=surgery
    index=result['long_title'].str.contains(surgery_name)
    result=surgery.loc[index[index.values==True].index,:]
    if download_as!='':
        result.to_csv(os.path.dirname(os.getcwd())+'/data/download data/'+download_as+'.csv', index=False)
    print('there are', len(result['long_title'].unique()), 'related surgery of', surgery_name, ':\n', 
          result['long_title'].unique())
    return result
```
