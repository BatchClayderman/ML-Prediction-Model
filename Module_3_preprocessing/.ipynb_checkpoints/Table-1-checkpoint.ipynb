```python
continuous_variables=['los','age','sofa','sapsii', 'heartrate_mean',
       'resprate_mean', 'tempc_mean', 'hemoglobin_mean', 'lactate_mean',
       'pco2_mean', 'ph_mean', 'po2_mean', 'potassium_mean','creatininecsv', 'glucosecsv',
       'neutrophilscsv', 'plateletcsv', 'potassiumcsv', 'sodiumcsv',
       'ureanitrogencsv', 'uric_acidcsv', 'wbccsv', 'albumin', 'map', 'bmi',
        'bilirubin','pt']
discrete_variable=['ethnicity', 'gender','first_careunit','hospital_mortality', '_mortality', 'v16',
       'v17','vasopressin', 'copdcsv', 'coronaryheartdiseasecsv', 'diabetescsv', 'hypertensioncsv',
       'liverdiseasecsv', 'aki']
```

```python
def times_count(data,variable_list,y):
    result=[]
    total=len(data)
    left_sum=len(data[data[y]==1])
    right_sum=len(data[data[y]==0])
    chi2_res=[]
    for variable in variable_list:
        for j in data[variable].unique():
            count1=len(data[(data[variable]==j)&(data[y]==1)])
            count0=len(data[(data[variable]==j)&(data[y]==0)])
            result.append([variable, j,count1,count1/total,count0,count0/total,str(variable)+str(j)])
    res_dataframe=pd.DataFrame(result,columns=['variable','value',y+'=1',y+'=1(%)',y+'=0',y+'=0(%)','id'])
    for variable in variable_list:
        for j in data[variable].unique():
            left_up=res_dataframe[(res_dataframe['variable']==variable)&(res_dataframe['value']==j)][y+'=1']
            right_up=res_dataframe[(res_dataframe['variable']==variable)&(res_dataframe['value']==j)][y+'=0']
            left_down=left_sum-left_up
            right_down=right_sum-right_up
            chi2_res.append([variable,j,stats.chi2_contingency(np.array([[left_up,right_up],[left_down,right_down]]))[1],str(variable)+str(j)])
    chi2_dataframe=pd.DataFrame(chi2_res,columns=['variable','value','chi2Test_p-value','id'])
    output=pd.merge(res_dataframe,chi2_dataframe[['chi2Test_p-value','id']],on='id',how='left')
    output.drop('id',axis=1,inplace=True)
    return output

def mean_and_std(data,variables,y):
    temp1=data[data[y]==1][variables].mean().to_frame(name=y+'_'+'mean'+'_1')
    temp2=data[data[y]==1][variables].std().to_frame(name=y+'_'+'std'+'_1')
    temp3=data[data[y]==0][variables].mean().to_frame(name=y+'_'+'mean'+'_0')
    temp4=data[data[y]==0][variables].std().to_frame(name=y+'_'+'std'+'_0')
    res1=pd.concat([temp1,temp2,temp3,temp4], axis=1)
    
    result=[]
    for i in variables:
        group1 = data[(data[y] == 1) & (data[i].notnull())][i]
        group2 = data[(data[y] == 0) & (data[i].notnull())][i]
        res = stats.mannwhitneyu(group1, group2)
        result.append([str(i),res[1]])
    res2=pd.DataFrame(result,columns=['variable','rank_sum_test_p-value'])
    res1['variable']=res1.index
    res=pd.merge(res1,res2,on='variable',how='left')
    return res.loc[:,['variable',y+'_'+'mean'+'_1', y+'_'+'std'+'_1',
                      y+'_'+'mean'+'_0',y+'_'+'std'+'_0', 'rank_sum_test_p-value']]
```

```python
discrete=times_count(afterPSM,discrete_variable,'group')
continuous=mean_and_std(afterPSM,continuous_variables,'group')
table_one=pd.concat([discrete,continuous],axis=1)
```
