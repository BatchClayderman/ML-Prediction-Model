```python
import pandas as pd
```

```python
data=pd.read_csv('data/merge data/biochemical indexes/creatinine.csv')
rrt=pd.read_csv('data/merge data/basic information/rrtfirstday.csv')
```

```python
data=data.sort_values(by=['subject_id','hadm_id','charttime'])
first_time=data.groupby('hadm_id').first()
max_creatinine=data.groupby('hadm_id')['valuenum'].max()
aki_data=pd.DataFrame({'hadm_id': first_time.index, 'first_value': first_time['valuenum'].values})
aki_data=pd.merge(aki_data, max_creatinine, on='hadm_id', how='left')
aki_data.columns=['hadm_id', 'first_value', 'max_value']
aki_data['aki']=0
aki_data.loc[1.5 * aki_data.first_value < aki_data.max_value, 'aki']=1
aki_data.loc[2 * aki_data.first_value < aki_data.max_value, 'aki']=2
aki_data.loc[3 * aki_data.first_value < aki_data.max_value, 'aki']=3
aki_data=pd.merge(aki_data, rrt[['hadm_id','rrt']], on='hadm_id', how='left')
aki_data.loc[aki_data[aki_data.rrt==1].index,'aki']=3
last_time=data.sort_values(by=['subject_id','hadm_id','charttime'],ascending=False).groupby('hadm_id').first()
last_value=pd.DataFrame({'hadm_id': last_time.index, 'last_value':last_time.valuenum.values})
aki_data=pd.merge(aki_data, last_value, on='hadm_id', how='left')
aki_data['aki_recover']=0
aki_data.loc[(aki_data['last_value'] < 1.5 * aki_data['first_value']) & aki_data['aki']!=0, 'aki_recover']=1
aki_data['aki_1']=0
aki_data['aki_2']=0
aki_data['aki_3']=0
aki_data.loc[aki_data['aki']==1, 'aki_1']=1
aki_data.loc[aki_data['aki']==2, 'aki_2']=1
aki_data.loc[aki_data['aki']==3, 'aki_3']=1
```

```python
aki_data.to_csv('data/original data/aki.csv', index = False)
```
