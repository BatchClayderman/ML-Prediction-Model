```python
import pandas as pd
```

```python
admission=pd.read_csv('C:/Users/IceFr/downloads/admissions.csv')
akf=pd.read_csv('C:/Programs/Python Project/data/merge data/disease/AKF.csv')
ESRD=pd.read_csv('C:/Programs/Python Project/data/merge data/disease/ESRD.csv')
```

```python
akf['akf']=1
ESRD['ESRD']=1
```

```python
admission=pd.merge(admission, akf[['hadm_id','akf']], on='hadm_id', how='left')
admission=pd.merge(admission, ESRD[['hadm_id','ESRD']], on='hadm_id', how='left')
```

```python
admission['akf']=admission['akf'].fillna(0)
```

```python
admission['ESRD']=admission['ESRD'].fillna(0)
```

```python
admission.sort_values(by=['hadm_id','admittime'],inplace=True)
```

```python
mark=admission.groupby('subject_id').count()['hadm_id']
```

```python
mark2=pd.DataFrame({'subject_id':mark.index, 'count':mark.values})
```

```python
target=mark2[mark2['count']!=1]
```

```python
target['target']=1
```

```python
admission=pd.merge(admission, target, on='subject_id', how='left')
```

```python
admission2=admission[admission['target']==1]
```

```python
admission2=admission2.sort_values(by=['subject_id', 'admittime'])
```
